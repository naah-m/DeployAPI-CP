## **Diagnóstico médico: Problemas Ortopédicos na Coluna Vertebral**

Esse dataset contém dados sobre problemas ortopédicos na coluna vertebral, diagnosticados no Centre Médico-Chirurgical de Réadaptation des Massues, em Lyon, France. Contém 6 atributos biomecânicos para 310 entradas anonimizadas, sendo 100 de pacientes considerados sem problemas (Normal - NO), 60 de pacientes com Hérnia de Disco (Disk Hernia - DH) e 150 de pacientes com Espondilolistese (Spondylolisthesis - SL).

Maiores informações disponíveis em https://www.openml.org/d/1523

Escolha dois algoritmos para treinar seus modelos de classificação.

Opções: KNN, Random Forest, Decision Tree, Logistic Regression, SVM

```python
# Criação do dataframe
import pandas as pd
from sklearn.datasets import fetch_openml

print("Baixando dados...")
dataset = fetch_openml(data_id=1523, as_frame=True, parser='auto')

df = dataset.frame

df.rename(columns={
    'V1': 'pelvic_incidence',
    'V2': 'pelvic_tilt',
    'V3': 'lumbar_lordosis_angle',
    'V4': 'sacral_slope',
    'V5': 'pelvic_radius',
    'V6': 'degree_spondylolisthesis',
    'Class': 'diagnostic'
}, inplace=True)

print("Dataframe criado!")
```

```python
# Visualize o tamanho do dataframe
print("Tamanho do DataFrame (linhas, colunas):", df.shape)
```

```python
# Visualize os primeiros registros
display(df.head())
```

```python
# Mostre uma tabela com as informações dos atributos (formato dos dados)
df.info()
```

```python
# Mostre uma tabela com as estatísticas descritivas dos atributos
display(df.describe())
```

```python
# Conversão de dados categóricos em numéricos (coluna "diagnostic")
mapa_diagnostico = {
    '1': 0,
    '2': 1,
    '3': 2
}
df['diagnostic'] = df['diagnostic'].map(mapa_diagnostico)
```

```python
# Visualização dos dados da coluna "diagnostic" após a transformação
print(df['diagnostic'].value_counts())
```

Crie as variáveis X e y.

X deve ser um dataframe que contenha os dados dos seis atributos referentes aos biomarcadores (V1 a V6).

y deve ser um array com os dados da coluna "diagnostic".

```python
# Coloque aqui o seu código para a variável X
X = df.drop(columns=['diagnostic'])
```

```python
# Coloque aqui o seu código para a variável y
y = df['diagnostic'].values
```

```python
# Visualize os primeiros registros do dataframe X
display(X.head())
```

```python
# Visualize X como um array Numpy (values)
X.values
```

```python
# Visualize y como um array Numpy
print(y)
```

Abaixo, separe 70% dos dados para treinamento e 30% para teste.

Utilize os nomes X_train, X_test, y_train e y_test


```python
# Separação de dados de treinamento e teste

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, random_state=42, stratify=y
)
```

```python
# Crie a instância do primeiro modelo:
modelo_rf = RandomForestClassifier(random_state=42, class_weight='balanced')
```

```python
# Treine o primeiro modelo:
modelo_rf.fit(X_train, y_train)
```

```python
# Gere o array de previsões "y_predict_1":
y_predict_1 = modelo_rf.predict(X_test)
```

```python
# Visualize y_predict_1:
print(y_predict_1)
```

```python
# Crie a instância do segundo modelo:

modelo_dt = DecisionTreeClassifier(random_state=42, class_weight='balanced')
```

```python
# Treine o segundo modelo:
modelo_dt.fit(X_train, y_train)
```

```python
# Gere o array de previsões "y_predict_2":
y_predict_2 = modelo_dt.predict(X_test)
```

```python
# Visualize y_predict_2:
print(y_predict_2)
```

```python
# Apresente a acurácia do primeiro modelo.
acc_rf = accuracy_score(y_test, y_predict_1)
print(f"Acurácia do primeiro modelo (Random Forest): {acc_rf * 100:.2f}%")
```

```python
# Apresente a acurácia do segundo modelo.
acc_dt = accuracy_score(y_test, y_predict_2)
print(f"Acurácia do segundo modelo (Decision Tree): {acc_dt * 100:.2f}%")
```

```python
#Criação e Exploração do DataFrame
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

df = pd.read_csv('/content/drive/MyDrive/FIAP/diabetes.csv')

print("Tamanho do DataFrame (linhas, colunas):", df.shape)

print("\nPrimeiros registros:")
display(df.head())

print("\nInformações dos atributos:")
df.info()

print("\nEstatísticas descritivas:")
display(df.describe())

print("\nContagem das classes na coluna 'Outcome':")
print(df['Outcome'].value_counts())
```

```python
#Criação das variáveis X e y
X = df.drop(columns=['Outcome'])
y = df['Outcome'].values

print("Primeiros registros de X:")
display(X.head())

print("\nVisualizando X como array Numpy (primeira linha):")
print(X.values[0])

print("\nVisualizando y como array Numpy (primeiros 15 registros):")
print(y[:15])
```

```python
#Treinamento e Teste
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, random_state=42, stratify=y
)

modelo_rf = RandomForestClassifier(random_state=42, class_weight='balanced')

modelo_rf.fit(X_train, y_train)

y_predict = modelo_rf.predict(X_test)

print("Previsões (y_predict):")
print(y_predict[:30])

acc_rf = accuracy_score(y_test, y_predict)
print(f"\nAcurácia do modelo (Random Forest): {acc_rf * 100:.2f}%")
```
